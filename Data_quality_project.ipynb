```python
# ML TASK: CLASSIFICATION
import pandas as pd

bank = pd.read_csv("bank.csv")
bank
```

```python
# importing dirty_completeness.py

import dirty_completeness as dc
df_list = []

# saving the dirt datasets
df_list = dc.injection(df_pandas=bank, seed=2, name="bank", name_class="y")
df_list
```

```python
# saving the dirt dataset with 50% completeness
df_dirt = df_list[1]
df_dirt
```

```python
df_dirt.isnull().sum()
```

```python
# Standard technique chosen: mode

# ho usato la moda perché è una tecnica che funziona sia nel caso in cui i valori mancanti sono stringhe, sia nel caso in cui sono valori numerici
# ho fatto la moda di ogni colonna, probabilmente c'era un metodo più veloce, però per ora non l'ho trovato

# fill NA/NaN values using the specified method
df_dirt["age"].fillna(df_dirt["age"].mode()[0])
```
